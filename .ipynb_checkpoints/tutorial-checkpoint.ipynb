```python
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
from IPython.display import display
import io
import warnings
import numpy as np
import pandas as pd
import nbinteract as nbi
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from scipy.interpolate import make_interp_spline
import math
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
# -*- coding: UTF-8 -*-

# widgets.Button(
#     description='Click me',
#     disabled=False,
#     button_style='success', # 'success', 'info', 'warning', 'danger' or ''
#     tooltip='Click me',
#     icon='check'
# )


# # 调整按钮
# from ipywidgets import Button, Layout

# b = Button(description='(50% width, 80px height) button',
#            layout=Layout(width='50%', height='80px'),
#           button_style='success')

# upload = widgets.FileUpload(
#     accept='',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
#     multiple=False,  # True to accept multiple files upload else False
# )
accordion = widgets.Accordion(children=[widgets.IntSlider(), widgets.Text()], titles=('Slider', 'Text'))

# widgets.HBox([b, upload])
tab_nest = widgets.Tab() # 标签
tab_nest.children = [accordion, widgets.Text(), widgets.Text(), widgets.Text()]
tab_nest.titles = ('air_vref_test', '1000ppm_test', "氧依赖性", "极限电流")
tab_nest.set_title(0, "特性曲线空气")
tab_nest.set_title(1, "特性曲线1000ppm")
tab_nest.set_title(2, "氧依赖性")
tab_nest.set_title(3, "极限电流")
tab_nest.titles = ("sdfsf", "sdfsdf")
tab_nest

upload_air = widgets.FileUpload(
    accept='.xlsx',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
    multiple=False,  # True to accept multiple files upload else False
)
def on_air_file_button_clicked(b):
    upload_air.disabled = False
    upload_air.value.clear()
    upload_air._counter = 0
    upload_air.data.clear()
    output._flush()
air_file_button = widgets.Button(description="重新选择")
air_test_air_data = widgets.HBox([widgets.Label(value="空气下联泵数据："), upload_air, air_file_button])
air_file_button.on_click(on_air_file_button_clicked)

upload_vref = widgets.FileUpload(
    accept='.xlsx',  # Accepted file extension e.g. '.txt', '.pdf', 'image/*', 'image/*,.pdf'
    multiple=False,  # True to accept multiple files upload else False
)
def on_vref_file_button_clicked(b):
    upload_vref.disabled = False
    upload_vref.value.clear()
    upload_vref._counter = 0
vref_file_button = widgets.Button(description="重新选择")
vref_test_air_data = widgets.HBox([widgets.Label(value="空气下Vref数据："), upload_vref, vref_file_button])
vref_file_button.on_click(on_vref_file_button_clicked)

picture_title = widgets.Text()
picture_title_data = widgets.HBox([widgets.Label(value="图片标题(为0默认为表格名称)："), picture_title])

x_title_lb = widgets.Label(value="横坐标(都为0则不限制)：")
x_min_title_lb = widgets.Label(value="最小值：")
x_min_title = widgets.Text()
x_max_title_lb = widgets.Label(value="最大值：")
x_max_title = widgets.Text()
x_axis = widgets.HBox([x_title_lb, x_min_title_lb, x_min_title, x_max_title_lb, x_max_title])

y_title_lb = widgets.Label(value="纵坐标(都为0则不限制)：")
y_min_title_lb = widgets.Label(value="最小值：")
y_min_title = widgets.Text()
y_max_title_lb = widgets.Label(value="最大值：")
y_max_title = widgets.Text()
y_axis = widgets.HBox([y_title_lb, y_min_title_lb, y_min_title, y_max_title_lb, y_max_title])

button = widgets.Button(description="生成")
output = widgets.Output()
display(air_test_air_data, vref_test_air_data, picture_title_data, x_axis, y_axis, button, output)

def read_excel(upload):
    uploaded_file = upload.value
    file_list = []
    data = {}
    for name in uploaded_file:
        file_list.append(name)
    with warnings.catch_warnings(record=True):
        warnings.simplefilter("always")
        data_xls = pd.ExcelFile(io.BytesIO(uploaded_file[file_list[0]]['content']))
        for name in data_xls.sheet_names:
            # df=data_xls.parse(sheet_name=name,header=None)
            df = data_xls.parse(sheet_name=name)
            data[name] = df
    return data

# 解析sheet数据
def handle_sheet_data(sheet_data, row_name, column_name, ip1_limit):
    try:
        index_data = sheet_data[sheet_data['IP2'].isin([row_name])].index.tolist()[0]
    except:
        index_data = -1
    index_x = sheet_data.columns.get_loc('温度') # 根据列名获取下标
    index_y = sheet_data.columns.get_loc(column_name) # 根据列名获取下标
    column_count = getTemperTime(sheet_data)
    if ip1_limit == True:
        ip1_index = find_ip1_close(sheet_data)
    else:
        ip1_index = 0
    x = sheet_data.values[index_data + 1 + ip1_index:index_data + 1 + column_count, index_x]
    y = sheet_data.values[index_data + 1 + ip1_index:index_data + 1 + column_count, index_y]
    data = {}
    data['x'] = x
    data['y'] = y
    return data

IP1_SET = 7
# 找到IP1闭环的下标
def find_ip1_close(sheet_data):
    index = 0
    try:
        index_data = sheet_data[sheet_data['IP2'].isin(['IP1'])].index.tolist()[0]
    except:
        index_data = -1
    index_ip1 = sheet_data.columns.get_loc('温度') + 1 # 根据列名获取下标 eg:V1_400
    column_count = getTemperTime(sheet_data)
    x = sheet_data.values[index_data + 1:index_data + 1 + column_count, index_ip1]
    for ip1 in x :
        ip1 = float(ip1)
        if ip1 >= (IP1_SET - 0.2) and ip1 <= (IP1_SET + 0.2):
            break
        index += 1
    return index

# series数据放入字典
def series_dic(sheet_data, paintName, seriesName, ip1_limit):
    data = {}
    for name in seriesName:
        data[name] = handle_sheet_data(sheet_data, paintName, name, ip1_limit)
    return data

# pic数据放入字典
def pic_dic(sheet_data, picNameList, seriesNameList, ip1_limit):
    data = {}
    for name in picNameList:
        data[name] = series_dic(sheet_data, name, seriesNameList, ip1_limit)
    return data
# 计算电导率，输入空气sheet，空泵sheet
def conductivity_data(air_sheet, none_sheet):
    data = {}
    data_kelvins = [] # 开尔文数组
    data_R = [] # 电阻数组
    data_x = []
    data_y = []
    air_dic = pic_dic(air_sheet, ['Vref', 'IP0'], ['V1_450'], False)
    none_dic = pic_dic(none_sheet, ['Vref'], ['V1_450'], False)
    air_vref_series = air_dic['Vref']['V1_450']
    air_ip0_series = air_dic['IP0']['V1_450']
    none_series = none_dic['Vref']['V1_450']
    for x in air_vref_series['x']:
        data_kelvins.append(x + 274.15) # 温度转开尔文
    for x in data_kelvins: # 1000 / T
        data_x.append(1000 / x)
    # 欧姆计算
    for air_vref, none_vref, air_ip0 in zip(air_vref_series['y'], none_series['y'], air_ip0_series['y']):
        try:
            data_R.append((none_vref - air_vref) / air_ip0 * 1000)
        except:
            data_R.append(0)
    # 电导率计算
    for R, T in zip(data_R, data_kelvins):
        try :
            data_y.append(math.log(1 / R * T))
        except:
            data_y.append(0)
        # data_y.append(0)
#         try:
#             print('电导率 ', 1 / R * T)
#         except:
#             print('电导率：0')
    data['x'] = np.array(data_x, dtype=object)
    data['y'] = np.array(data_y, dtype=object)
    data_1 = {}
    data_2 = {}
    data_1['1000T ln(1/R*T)'] = data
    # T R
    data_11 = {}
    data_11['x'] = np.array(data_kelvins, dtype=object)
    data_11['y'] = np.array(data_R, dtype=object)
    data_2['T R'] = data_11
    returnData = {}
    returnData['1000T ln(1/R*T)'] = data_1
    returnData['T R'] = data_2
    return returnData

# 获取气氛数量
def getTemperTime(sheet_data):
    time = 0 # 气氛数量
    count = sheet_data.columns.get_loc("序号") # 根据列名获取下标
    count_index = 0 # 序号的索引
    data = sheet_data.values[count_index:, count] # 数据
    for index in range(1, len(data)): # 从1开始去掉序号
        if math.isnan(data[index]): # 找到nan位置，再减1
            time = index
            break
        else:
            time = index
    return time

def getTotalValue(pic):
    TotalValue = np.array([])
    for picname in pic: # 遍历系列数据
        series = pic[picname]
        TotalValue = np.append(TotalValue, series['y'])
    try:
        print('Max', TotalValue.max())
    except:
        print()
    # print('Max', TotalValue.max())
    print('tatal', TotalValue)
    return TotalValue

# 'upper right'  : 1,
# 'upper left'   : 2,
# 'lower left'   : 3,
# 'lower right'  : 4,
# 'right'        : 5,
# 'center left'  : 6,
# 'center right' : 7,
# 'lower center' : 8,
# 'upper center' : 9,
# 'center'       : 10
inset_index = [9, 2, 3, 8, 4, 1]

# 获取系列名称
def getSeriesTitle(sheet_data):
    column = sheet_data.columns.tolist()
    column.remove('V2')
    column.remove('序号')
    column.remove('温度')
    column.remove('IP2')
    print(column)
    return column

# series_name_list = ['V1_400', 'V1_410', 'V1_420', 'V1_430', 'V1_440', 'V1_450']
pic_name_list = ['IP2', 'VP0', 'Vref', 'IP0']


def handledata(two_excel):
    locate = 1
    plt.style.use('bmh')
    fig = plt.figure(figsize=(60, 40), facecolor="white")
    for air_test, none_test in zip(two_excel['air_test'], two_excel['none_test']): # 遍历sheet数据
        sub_pic = fig.add_subplot(8, 4, locate, facecolor="white") # 添加子控件
        sub_pic.set_xticks([]) # 横纵坐标去除
        sub_pic.set_yticks([]) # 纵纵坐标去除
        sub_pic.set_title("V5V5V7改温场--特性曲线(空气)0908 " + air_test, y = 1.05) # 标题
        series_name_list = getSeriesTitle(two_excel['air_test'][air_test])
        conductivity_pic = conductivity_data(two_excel['air_test'][air_test], two_excel['none_test'][none_test])
        pic_data = pic_dic(two_excel['air_test'][air_test], pic_name_list, series_name_list, True) # 获取需要显示的图
        pic_data['1000T ln(1/R*T)'] = conductivity_pic['1000T ln(1/R*T)']
        pic_data['T R'] = conductivity_pic['T R']
        # print(pic_data)
        index = 0
        for name in pic_data: # 遍历图片数据
            pic = pic_data[name]
            axins1 = inset_axes(sub_pic,width="30%", height="40%", loc=inset_index[index]) # 设置显示区域大小
#             total_value = getTotalValue(pic)
            for picname in pic: # 遍历系列数据
                series = pic[picname]
                if index < 4:
                    try: 
                        x_smooth = np.linspace(series['x'].min(), series['x'].max(), 12) #100 represents number of points to make between T.min and T.max 拟合曲线
                        y_smooth = make_interp_spline(series['x'], series['y'])(x_smooth)
                    except:
                        x_smooth = series['x']
                        y_smooth = series['y']
                    axins1.plot(x_smooth, y_smooth, linewidth=1, label=picname) # 画图
                    # try:
                    #     max = total_value.max()
                    #     min = total_value.min()
                    # except:
                    #     max = 100
                    #     min = 0
                    # print('max', max, 'min', min)
                    # axins1.set_ylim(min, max)
                    try: # 避免横坐标为0
                        axins1.set_xlim(series['x'].min(), series['x'].max())
                    except:
                        pass
                else:
                    axins1.plot(series['x'], series['y'], linewidth=1, label=picname) # 画图
            # axins1.set_ylim(pic_ylim[index - 1])
            # 

            axins1.set_title(name, color='blue')
            # axins1.set_xlabel("T") #xlabel、ylabel：分别设置X、Y轴的标题文字。
            # axins1.style.use('bmh')
            axins1.legend()
            index = index + 1
        # if locate == 2 or locate == 7:
        #     locate = locate + 1
        locate = locate + 1

#     plt.savefig('V5V5V7改温场--特性曲线(空气)0908.png',dpi=200,bbox_inches='tight')
    plt.show()

def on_button_clicked(b):
    two_excel = {}
    two_excel["air_test"] = read_excel(upload_air) # 空气下联泵数据
    two_excel["none_test"] = read_excel(upload_vref) # 空泵vref
    delSheetName = ['zongbiao', 'v1step', 'unitpump', 'firstpump', 'secondpump', 'thirdpump', 'O2Dependency']
    for data in two_excel:
        for sheetName in list(two_excel[data].keys()): # 字典遍历时不能删除，将字典转化为列表
            if sheetName in delSheetName:
                try:
                    del two_excel[data][sheetName]
                except:
                    pass
    print(two_excel[data].keys())
    print(two_excel['air_test']['1_NGK28'].columns.str.strip())
    print(two_excel['air_test']['1_NGK28'].columns.tolist().index('V1_420'))
    series = getSeriesTitle(two_excel['air_test']['1_NGK28'])
    handledata(two_excel)

count = 0
# upload.on_incr_counter(count)
button.on_click(on_button_clicked)
# layout = widgets.VBox([button, output])
# layout
```
